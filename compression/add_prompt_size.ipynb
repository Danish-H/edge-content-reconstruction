```python
import json

# Sample data
with open('bws_gpt4_color2.json') as f:
    jsons = json.load(f)

# Example dictionary of prompts with sizes (you can calculate these sizes based on actual text)
prompts = prompts_gpt4
# for each key change it to k.split(".")[0])
prompts = {k.split(".")[0]: v for k, v in prompts.items()}

# Function to calculate the size of a text prompt
def calculate_prompt_size(prompt_text):
    return len(prompt_text.encode('utf-8'))  # Size in bytes for UTF-8 encoded string

# Update the JSON objects
for image_json in jsons:
    image_number = str(image_json["image no"]).zfill(3)  # Convert number to string and pad with zeros
    if image_number in prompts:
        prompt_size = calculate_prompt_size(prompts[image_number])
        image_json["prompt size"] = prompt_size
        image_json["post compression size with prompt"] = image_json["post compression size"] + prompt_size  # Add the prompt size to the post-compression size

# Print updated JSONs
print(json.dumps(jsons, indent=4))

# Save json
with open('bws_gpt4_color2.json', 'w') as f:
    json.dump(jsons, f, indent=4)
```
