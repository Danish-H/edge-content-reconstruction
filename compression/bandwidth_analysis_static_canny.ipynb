```python
import os

# Paths to the directories
dataset_dir = '..\images_dataset'
canny_dataset_dir = '..\images_dataset_canny'

# Get the list of image file names in each directory
dataset_images = sorted(os.listdir(dataset_dir))
canny_dataset_images = sorted(os.listdir(canny_dataset_dir))

# Check if directories contain the same number of files
if len(dataset_images) != len(canny_dataset_images):
    print("Warning: The directories do not contain the same number of images.")

# Full paths to the images
dataset_image_paths = [os.path.join(dataset_dir, img) for img in dataset_images]
canny_dataset_image_paths = [os.path.join(canny_dataset_dir, img) for img in canny_dataset_images]

# Print the lists (optional)
print("Dataset Images:", dataset_image_paths)
print("Canny Dataset Images:", canny_dataset_image_paths)
```

```python
# Get file sizes in bytes for each image
dataset_image_sizes = [os.path.getsize(path) for path in dataset_image_paths]

# Print the file sizes (optional)
for img, size in zip(dataset_images, dataset_image_sizes):
    print(f"{img}: {size} bytes")
```

```python
import subprocess
import time

canny_dataset_images_sizes = []

for i in range(150):
    result = subprocess.run(['.\jbig2', '-s', '-p', canny_dataset_image_paths[i]], text=True, capture_output=True)
    time.sleep(0.3)
    canny_dataset_images_sizes.append(os.path.getsize('./output.0000') + os.path.getsize('./output.sym'))
```

```python
bandwidth_savings_percentages = []

for i in range(150):
    saving = (dataset_image_sizes[i] - canny_dataset_images_sizes[i]) / dataset_image_sizes[i]
    saving = saving * 100
    bandwidth_savings_percentages.append(saving)
```

```python
# print(bandwidth_savings_percentages)

print("Minimum bandwidth saving (percentage):", min(bandwidth_savings_percentages), "%")
print("Maximum bandwidth saving (percentage):", max(bandwidth_savings_percentages), "%")
print("Mean bandwidth saving (percentage):", sum(bandwidth_savings_percentages) / len(bandwidth_savings_percentages), "%")
```

```python
import json

# List to hold all the data dictionaries
data = []

# Populate the list with dictionaries
for i in range(150):
    entry = {
        'image no': i+1,
        'original image size': dataset_image_sizes[i],
        'post compression size': canny_dataset_images_sizes[i],
        'bandwidth savings percentage': bandwidth_savings_percentages[i]
    }
    data.append(entry)

# Write the data to a JSON file
with open('bandwidth_savings_static_canny.json', 'w') as f:
    json.dump(data, f, indent=4)

print("JSON file has been created and saved.")
```
